```python
# Importer la bibliothèque pandas
import pandas as pd

# Créer un dictionnaire pour stocker les données des moyennes mobiles par année
mm_data = {}

# Boucler à travers chaque année
for year in range(2012, 2015):  # Nous avons des données pour les années 2012, 2013 et 2014
    # Filtrer les données pour l'année spécifique
    data_year = data.loc[f'{year}-01-01':f'{year}-12-31']

    # Sélectionner les prix de clôture pour Porsche et Volkswagen pour cette année
    porsche_close_year = data_year["PAH3.DE"]["Close"]
    volkswagen_close_year = data_year["VOW3.DE"]["Close"]

    # Calculer la moyenne mobile simple sur une fenêtre de 10 jours pour Porsche
    porsche_sma_10 = porsche_close_year.rolling(window=10).mean()

    # Calculer la moyenne mobile simple sur une fenêtre de 10 jours pour Volkswagen
    volkswagen_sma_10 = volkswagen_close_year.rolling(window=10).mean()

    # Stocker les résultats dans le dictionnaire mm_data
    mm_data[f'{year}'] = {'Porsche': porsche_sma_10, 'Volkswagen': volkswagen_sma_10}

# Afficher les résultats pour chaque année
for year, values in mm_data.items():
    print(f"Moyenne mobile pour l'année {year}:")
    print("Porsche:\n", values['Porsche'])
    print("Volkswagen:\n", values['Volkswagen'])
    print("\n")
```

# **La tendance générale du prix de l'action de Porsche et volkswagen**

```python
import matplotlib.pyplot as plt

# Filtrer les données pour les années 2010 et 2013
data_2010 = data.loc['2010-01-01':'2010-12-31']
data_2013 = data.loc['2013-01-01':'2013-12-31']

# Sélectionner les prix de clôture pour Porsche et Volkswagen pour chaque année
porsche_close_2010 = data_2010["PAH3.DE"]["Close"]
volkswagen_close_2010 = data_2010["VOW3.DE"]["Close"]

porsche_close_2013 = data_2013["PAH3.DE"]["Close"]
volkswagen_close_2013 = data_2013["VOW3.DE"]["Close"]

# Créer une figure et des sous-graphiques
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Tracer les actions de Porsche et Volkswagen pour l'année 2010
axes[0, 0].plot(porsche_close_2010, label='Porsche')
axes[0, 0].plot(volkswagen_close_2010, label='Volkswagen')
axes[0, 0].set_title('Actions de Porsche et Volkswagen en 2010')
axes[0, 0].legend()

# Tracer les actions de Porsche et Volkswagen pour l'année 2013
axes[0, 1].plot(porsche_close_2013, label='Porsche')
axes[0, 1].plot(volkswagen_close_2013, label='Volkswagen')
axes[0, 1].set_title('Actions de Porsche et Volkswagen en 2013')
axes[0, 1].legend()

# Calculer si la valeur a augmenté ou diminué
porsche_diff_2010 = porsche_close_2010.iloc[-1] - porsche_close_2010.iloc[0]
volkswagen_diff_2010 = volkswagen_close_2010.iloc[-1] - volkswagen_close_2010.iloc[0]

porsche_diff_2013 = porsche_close_2013.iloc[-1] - porsche_close_2013.iloc[0]
volkswagen_diff_2013 = volkswagen_close_2013.iloc[-1] - volkswagen_close_2013.iloc[0]

# Créer des barres pour indiquer l'augmentation ou la diminution de la valeur
axes[1, 0].bar(['Porsche', 'Volkswagen'], [porsche_diff_2010, volkswagen_diff_2010], color=['green' if diff >= 0 else 'red' for diff in [porsche_diff_2010, volkswagen_diff_2010]])
axes[1, 0].set_title('Augmentation/Diminution de la valeur en 2010')

axes[1, 1].bar(['Porsche', 'Volkswagen'], [porsche_diff_2013, volkswagen_diff_2013], color=['green' if diff >= 0 else 'red' for diff in [porsche_diff_2013, volkswagen_diff_2013]])
axes[1, 1].set_title('Augmentation/Diminution de la valeur en 2013')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
# Supposez que vous avez déjà les données des prix de clôture pour Porsche et Volkswagen dans les variables porsche_close et volkswagen_close

# Calcul des rendements des actions
porsche_returns = porsche_close_2013.pct_change().dropna()  # Suppression des valeurs manquantes
volkswagen_returns = volkswagen_close_2013.pct_change().dropna()  # Suppression des valeurs manquantes

# Calcul de la corrélation entre les rendements
correlation = porsche_returns.corr(volkswagen_returns)

print("Corrélation entre les rendements des actions de Porsche et Volkswagen :", correlation)
```

**Une corrélation de 0.83 entre les rendements des actions de Porsche et Volkswagen** indique une forte corrélation positive. Cela suggère que les rendements des deux actions sont étroitement liés et qu'ils évoluent généralement dans la même direction. Cette forte corrélation pourrait être interprétée comme une indication de synergie entre les deux entreprises, bien qu'une analyse plus approfondie soit nécessaire pour confirmer cette conclusion.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supposons que vous avez déjà les données des prix de clôture pour Porsche et Volkswagen

# Créer un DataFrame pour stocker les moyennes annuelles des prix de clôture
annual_trend = pd.DataFrame(index=range(2010, 2016), columns=['Porsche', 'Volkswagen'])

# Boucle sur chaque année de 2010 à 2015
for year in range(2010, 2016):
    # Filtrer les données pour l'année spécifique
    data_year = data.loc[f'{year}-01-01':f'{year}-12-31']

    # Sélectionner les prix de clôture pour Porsche et Volkswagen pour cette année
    porsche_close_year = data_year["PAH3.DE"]["Close"]
    volkswagen_close_year = data_year["VOW3.DE"]["Close"]

    # Calculer la moyenne des prix de clôture pour cette année et les stocker dans le DataFrame
    annual_trend.loc[year, 'Porsche'] = porsche_close_year.mean()
    annual_trend.loc[year, 'Volkswagen'] = volkswagen_close_year.mean()

# Tracer la courbe de tendance annuelle
plt.figure(figsize=(10, 6))

# Tracer la courbe de tendance annuelle pour Porsche
plt.plot(annual_trend.index, annual_trend['Porsche'], marker='o', linestyle='-', label='Porsche')

# Tracer la courbe de tendance annuelle pour Volkswagen
plt.plot(annual_trend.index, annual_trend['Volkswagen'], marker='o', linestyle='-', label='Volkswagen')

# Ajouter des titres et légendes
plt.title('Courbe de tendance annuelle des actions Porsche et Volkswagen (2010-2015)')
plt.xlabel('Année')
plt.ylabel('Prix de clôture moyen')
plt.legend()
plt.grid(True)

# Afficher le graphique
plt.show()
```
